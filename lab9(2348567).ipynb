```python
import numpy as np
import matplotlib.pyplot as plt

bias = 5
weights = 2
input_values = np.linspace(4, 8, 100)

def unit_step(x):
    return np.where(x >= 0, 1, 0)

def sign_function(x):
    return np.sign(x)

def linear_function(x):
    return x

def piecewise_linear_function(x):
    return np.piecewise(x, [x < 5, (x >= 5) & (x < 7), x >= 7], [0, lambda x: x - 5, 2])

def logistic_function(x):
    return 1 / (1 + np.exp(-x))

def tanh_function(x):
    return np.tanh(x)

# Calculate neuron outputs
neuron_output_unit_step = unit_step(weights * input_values + bias)
neuron_output_sign = sign_function(weights * input_values + bias)
neuron_output_linear = linear_function(weights * input_values + bias)
neuron_output_piecewise_linear = piecewise_linear_function(weights * input_values + bias)
neuron_output_logistic = logistic_function(weights * input_values + bias)
neuron_output_tanh = tanh_function(weights * input_values + bias)

# Plot the results
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(input_values, neuron_output_unit_step, label='Unit Step')
plt.title('Unit Step Activation Function')

plt.subplot(2, 3, 2)
plt.plot(input_values, neuron_output_sign, label='Sign (Signum)')
plt.title('Sign Activation Function')

plt.subplot(2, 3, 3)
plt.plot(input_values, neuron_output_linear, label='Linear')
plt.title('Linear Activation Function')

plt.subplot(2, 3, 4)
plt.plot(input_values, neuron_output_piecewise_linear, label='Piece-wise Linear')
plt.title('Piece-wise Linear Activation Function')

plt.subplot(2, 3, 5)
plt.plot(input_values, neuron_output_logistic, label='Logistic (Sigmoid)')
plt.title('Logistic Activation Function')

plt.subplot(2, 3, 6)
plt.plot(input_values, neuron_output_tanh, label='Hyperbolic Tangent')
plt.title('Hyperbolic Tangent Activation Function')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Perceptron function
def perceptron(inputs, weights):
    # Add bias term (input 0 with weight w0)
    inputs = np.insert(inputs, 0, 1)
    # Compute the weighted sum
    weighted_sum = np.dot(inputs, weights)
    # Apply step function as activation
    return 1 if weighted_sum > 0 else 0

# Perceptron weights
weights = np.array([0.2, 0.5, 0.243])

# AND operation
inputs_and = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
outputs_and = [perceptron(inputs, weights) for inputs in inputs_and]
error_and = np.array([0, 1, 0, 1]) - np.array(outputs_and)

# OR operation
inputs_or = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
outputs_or = [perceptron(inputs, weights) for inputs in inputs_or]
error_or = np.array([0, 1, 1, 1]) - np.array(outputs_or)

# XOR operation
inputs_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
outputs_xor = [perceptron(inputs, weights) for inputs in inputs_xor]
error_xor = np.array([0, 0, 0, 1]) - np.array(outputs_xor)

# Display results
print("AND Operation:")
print("Inputs: ", inputs_and)
print("Outputs:", outputs_and)
print("Error:  ", error_and)

print("\nOR Operation:")
print("Inputs: ", inputs_or)
print("Outputs:", outputs_or)
print("Error:  ", error_or)

print("\nXOR Operation:")
print("Inputs: ", inputs_xor)
print("Outputs:", outputs_xor)
print("Error:  ", error_xor)
```
