```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_classifier(X_train, X_test, y_train, y_test, classifier):
    print("Fitting the model...")
    classifier.fit(X_train, y_train)
    print("Model fitted successfully.")

    print("Making predictions...")
    y_pred = classifier.predict(X_test)
    print("Predictions made successfully.")

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    print("Performance metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print()

# Load your dataset
iris = pd.read_csv("/content/IRIS.csv")

# Assume 'features' are all columns except the target variable, and 'target' is the target variable column
features = iris.drop('species', axis=1)  # Assuming 'species' is the target variable column
target = iris['species']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create an instance of the classifier
classifier = RandomForestClassifier()

# Call the evaluate_classifier function
evaluate_classifier(X_train, X_test, y_train, y_test, classifier)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Apply LDA for dimensionality reduction
lda = LDA()
X_train_lda = lda.fit_transform(X_train, y_train)
X_test_lda = lda.transform(X_test)

# Define classifiers
knn_classifier = KNeighborsClassifier()
decision_tree_classifier = DecisionTreeClassifier()
random_forest_classifier = RandomForestClassifier()

# Evaluate classifiers after dimensionality reduction
evaluate_classifiers(X_train_pca, X_test_pca, y_train, y_test, knn_classifier, "KNN (PCA - Iris)")
evaluate_classifiers(X_train_pca, X_test_pca, y_train, y_test, decision_tree_classifier, "Decision Tree (PCA - Iris)")
evaluate_classifiers(X_train_pca, X_test_pca, y_train, y_test, random_forest_classifier, "Random Forest (PCA - Iris)")

evaluate_classifiers(X_train_lda, X_test_lda, y_train, y_test, knn_classifier, "KNN (LDA - Iris)")
evaluate_classifiers(X_train_lda, X_test_lda, y_train, y_test, decision_tree_classifier, "Decision Tree (LDA - Iris)")
evaluate_classifiers(X_train_lda, X_test_lda, y_train, y_test, random_forest_classifier, "Random Forest (LDA - Iris)")
```

```python
# Apply LDA for dimensionality reduction
lda_diab = LDA(n_components=1)  # Change n_components to 1
X_train_diab_lda = lda_diab.fit_transform(X_train_diab, y_train_diab)
X_test_diab_lda = lda_diab.transform(X_test_diab)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_train_diab_pca[:, 0], X_train_diab_pca[:, 1], c=y_train_diab, cmap='viridis', edgecolor='k', s=40)
plt.title('PCA - Reduced Dimensions')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.subplot(1, 2, 2)
plt.scatter(X_train_diab_lda, y_train_diab, c=y_train_diab, cmap='viridis', edgecolor='k', s=40)
plt.title('LDA - Reduced Dimensions')
plt.xlabel('Linear Discriminant 1')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Visualize PCA
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(range(1, pca_diab.n_components_ + 1), pca_diab.explained_variance_ratio_, alpha=0.8, align='center')
plt.title('Explained Variance Ratio - PCA')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.xticks(range(1, pca_diab.n_components_ + 1))

# Visualize LDA
plt.subplot(1, 2, 2)
lda_diab_coef = lda_diab.scalings_[:, :lda_diab.n_components]
plt.bar(range(1, lda_diab.n_components + 1), np.abs(lda_diab_coef.squeeze()), alpha=0.8, align='center')
plt.title('Absolute Coefficients - LDA')
plt.xlabel('Linear Discriminant')
plt.ylabel('Absolute Coefficient')
plt.xticks(range(1, lda_diab.n_components + 1))

plt.tight_layout()
plt.show()
```
